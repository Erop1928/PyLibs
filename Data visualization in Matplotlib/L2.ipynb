## Задание 2

```python
t = np.linspace(51, 0, 10)
f = np.cos(t)

plt.plot(t, f, color="green")
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.title("График f(t)")
plt.xlabel("Значения t")
plt.ylabel("Значения f")
```

## Задание 3*

```python
x = np.linspace(51, -3, 3)

y1 = x**2
y2 = x * 2 + 0.5
y3 = x * (-3) - 1.5
y4 = np.sin(x)

fig, axs = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = axs.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title("График y1")

ax2.plot(x, y2)
ax2.set_title("График y2")

ax3.plot(x, y3)
ax3.set_title("График y3")

ax4.plot(x, y4)
ax4.set_title("График y4")

plt.show()
```

## Задание 4*

```python
import pandas as pd
import matplotlib.style as style

style.use('fivethirtyeight')

credit_table = pd.read_csv("creditcard.csv")

val_counts = credit_table.Class.value_counts()
val_counts.plot(kind="bar")

plt.show()
```

```python
val_counts.plot(kind="bar", logy=True)

plt.show()
```

```python
plt.hist(credit_table[credit_table.Class==1].V1, density=True, bins=20, alpha=0.5, color='red')
plt.hist(credit_table[credit_table.Class==0].V1, density=True, bins=20, alpha=0.5, color='gray')
plt.xlabel("Cl")
plt.legend(labels=["Class 1", "Class 0"])
plt.show()
```
